### <font color=darkred>Introduction</font>

The purpose of this exercise is recreate the following  Wikipedia Box Plot by producing multiple box plots from a dataframe on the one plot

![Wikipedia Image](https://upload.wikimedia.org/wikipedia/commons/thumb/f/fa/Michelsonmorley-boxplot.svg/800px-Michelsonmorley-boxplot.svg.png)

```python
# Importing Libraries
import numpy as np
import matplotlib.pyplot as plt
```

Firstly, for this exercise, I needed to read in the dataset from the 'data' folder in Jupyter using numpy. This can also be done using using pandas but for this exercise Numpy is used. As Numpy cannot remove the 'headings' rows when the dataset is being read in the <a href="https://numpy.org/doc/stable/reference/generated/numpy.genfromtxt.html" target="_top">skip_header</a> parameter is used to remove the first row from the dataset. The  <a href="https://numpy.org/doc/stable/reference/generated/numpy.genfromtxt.html" target="_top">comma delimiter</a> is used to seperate the values in the dataset.

```python
# Read in the morley data using Numpy
# remove the header, seperate using commas 
data = np.genfromtxt('data/morley.csv', skip_header=1, delimiter=',')
```

Next, as we don't need the first column of data (column index 0) in the dataframe it can be removed <a href="https://stackoverflow.com/questions/34007632/how-to-remove-a-column-in-a-numpy-array" target="_top">stackoverflow, (2022).</a>

```python
# Removing comuln index 0 from the dataframe 
data = data [:,1:]
#data
```

The next step is to plot the boxplots of data on one plot. The data we are interested in is the data in column index 2 and all of the rows of this column. Taking all the data from column index 2 and all of the rows in column 2 using <a href="https://www.w3schools.com/python/numpy/numpy_array_slicing.asp" target="_top">Numpy Array Slicing </a>. Next, reshape the array to an array of 5 columns of 20 (which is the 5 experiments each containing 20 runs) using the <a href="https://www.w3schools.com/python/numpy/numpy_array_reshape.asp" target="_top">reshape()</a> function. Lastly, to change the arrays from rows format to column format using <a href="https://www.geeksforgeeks.org/python-numpy-ndarray-t/" target="_top">Transpose (.T)</a>.


```python
# Reshaping the dataframe
data = data[:,2].reshape(5, 20).T #taking data from all of the 2nd index column reshaping it 
#data
```

Lastly, using code from <a href="https://stackoverflow.com/questions/50555179/how-to-add-data-labels-to-boxplot" target="_top">stackoverflow, (2022) </a>, I have generated a plot with the 5 boxplots plotted.

```python
# Creating the figure 
fig = plt.figure() 
ax = fig.add_subplot(111) #to adjust the size of the boxplot
ax.set_title('Boxplot') #title of the boxplot 
ax.set_xlabel('Experiment Number') #labeling the x axis 
ax.set_ylabel('Speed of light (km/s minus 299,000)') #labeling the x axis 
ax.grid() #Adding a grid
ax.boxplot(data, widths = 0.5, patch_artist = True); #printing the plot calling in the data, the width of the box plots and adding colour to the plots
```
