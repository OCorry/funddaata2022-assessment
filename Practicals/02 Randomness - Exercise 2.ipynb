
# <font color =green >Why are there the same number of ways to get 4 tails as there are to get 4 heads?</font>



There are the same number of ways (combinations) of getting 4 tails as there are 4 heads when flipping 10 coins. Fig.1 below demonstrates that having flipped 10 coins, if 4 are heads, then the remaining 6 must be tails and likewise, if there are 4 tails, the remaining 6 must be heads as there is no other possible outcomes. 

The number of possible combinations for 4 heads from flipping 10 coins is calculated: `(10 * 9 * 8 * 7) // (4 * 3 * 2 * 1)` , then the same formula applies to calculate the number of possible combinations of 4 tails. This is because, once the 4 placeholders are filled with heads, the remaining 6 placeholders must be filled with tails and likewise when placeholders are filled with tails, the remaining 6 placeholders must be occupied by heads (to allow for 10 successfully flipped coins).

```python
import numpy as np 
import matplotlib.pyplot as plt

x = np.array([10, 0])
y = np.array([0, 10])
plt.plot(x,y, 'g', label = "Total Heads+Tails", marker = 's')

font1 = {'color':'red','size':15} #https://www.w3schools.com/python/matplotlib_labels.asp 25/10/2022
font2 = {'color' : 'darkblue', 'size':15} 
plt.title("Fig:1: 10 Flipped Coins", fontdict =font2)
plt.ylabel("Number of Heads",fontdict = font1) #https://www.w3schools.com/python/matplotlib_labels.asp 25/10/2022
plt.xlabel("Number of Tails",fontdict = font1)
plt.xticks(np.arange(0, 11, 1)) #https://www.geeksforgeeks.org/python-matplotlib-pyplot-ticks/ 25/10/2022
plt.yticks(np.arange(0, 11, 1))
plt.grid()
plt.legend() #https://www.geeksforgeeks.org/custom-legends-with-matplotlib/
plt.show()
```
