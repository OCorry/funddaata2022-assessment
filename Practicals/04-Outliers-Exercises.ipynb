## <font color=darkred>Exercise 1-Recreating a Wikipedia Box Plot</font>
***

### <font color=darkred>Introduction</font>

The purpose of this exercise is recreate the following  <a href="https://en.wikipedia.org/wiki/Box_plot#/media/File:Michelsonmorley-boxplot.svg" target="_top">Wikipedia Box Plot</a> by producing multiple box plots from a dataframe on the one plot.

***


```python
# Importing Libraries
import numpy as np
import matplotlib.pyplot as plt
```

Firstly, for this exercise, I needed to read in the dataset from the 'data' folder in Jupyter using Numpy. This can also be done using using pandas but for this exercise Numpy is used. As Numpy cannot remove the 'headings' rows when the dataset is being read in the <a href="https://numpy.org/doc/stable/reference/generated/numpy.genfromtxt.html" target="_top">skip_header</a> parameter is used to remove the first row from the dataset. The  <a href="https://numpy.org/doc/stable/reference/generated/numpy.genfromtxt.html" target="_top">comma delimiter</a> is used to seperate the values in the dataset.

```python
# Read in the morley data using Numpy
# remove the header, seperate using commas 
data = np.genfromtxt('data/morley.csv', skip_header=1, delimiter=',')
```

Next, as we don't need the first column of data (column index 0) in the dataframe it can be removed <a href="https://stackoverflow.com/questions/34007632/how-to-remove-a-column-in-a-numpy-array" target="_top">stackoverflow, (2022).</a>

```python
# Removing column index 0 from the dataframe 
data = data [:,1:]
#data
```

The next step is to plot the boxplots of data on one plot. The data we are interested in is the data in column index 2 and all of the rows of this column. Taking all the data from column index 2 and all of the rows in column 2 using <a href="https://www.w3schools.com/python/numpy/numpy_array_slicing.asp" target="_top">Numpy Array Slicing </a>. Next, reshape the array to an array of 5 columns of 20 (which is the 5 experiments each containing 20 runs) using the <a href="https://www.w3schools.com/python/numpy/numpy_array_reshape.asp" target="_top">reshape()</a> function. Lastly, to change the arrays from rows format to column format using <a href="https://www.geeksforgeeks.org/python-numpy-ndarray-t/" target="_top">Transpose (.T)</a>.


```python
# Reshaping the dataframe
data = data[:,2].reshape(5, 20).T #taking data from all of the 2nd index column reshaping it 
#data
```

Lastly, using code from <a href="https://stackoverflow.com/questions/50555179/how-to-add-data-labels-to-boxplot" target="_top">stackoverflow, (2022) </a>, I have generated a plot with the 5 boxplots plotted.

```python
# Creating the figure 
fig = plt.figure() 
ax = fig.add_subplot(111) #to adjust the size of the boxplot
ax.set_title('Boxplot') #title of the boxplot 
ax.set_xlabel('Experiment Number') #labeling the x axis 
ax.set_ylabel('Speed of light (km/s minus 299,000)') #labeling the x axis 
ax.grid() #Adding a grid
ax.boxplot(data, widths = 0.5, patch_artist = True); #printing the plot calling in the data, the width of the box plots and adding colour to the plots
```

### <font color=darkred>Exercise 3-Adapt Ian's code and plots so that the overall plot is inversely proportional and the individual groups are directly proportional</font>
***

### <font color =darkred> Introduction</font>
The goal for this exercise is to adapt code from the Outliers section of <a href="https://github.com/ianmcloughlin/2223-S1-fund-data-analysis/blob/main/notebooks/04-outliers.ipynb" target="_top">McLoughlin. I, (2022)</a> Jupyter Notebook so that the overall plot is inversely proportional and the individual group are directly proportional.
***

```python
# Create a series of x vectors.
xs = np.array([np.linspace(i, i + 2.0, 20) for i in range(10)])
```

```python
# Create a corresponding series of y vectors.
ys = np.array([-1.0 * xs[i] + 2 * i for i in range(len(xs))])

# Add some noise.
ys = ys + np.random.normal(0.0, 1.0, ys.shape)
```

```python
# Create global list of x and y values.
x = xs.flatten()
y = ys.flatten()
```

To change the plot so that the overall plot is inversely proportional, I changed the *y* variable below from a positive to a negative. So now, as the value for **x** goes up, the value for **y** goes down. This is demonstrated in **Plot 1** below:

```python
# Plot
plt.plot(x, -y, '.');
plt.title('Plot 1: Overall Plot inversely proportional', color = "blue");
```

### <ins>Best Fit Line</ins>

```python
# Best fit line slope and y intercept.
m, c = np.polyfit(x, -y, 1)
```

```python
# Create a steadily increasing x range for plotting.
xbestfit = np.linspace(x.min(), x.max(), 1000)
# Create corresponding y values form best fit line.
ybestfit = m * xbestfit + c
```

Again, to demonstrate the *best fit line* I adaped the code by changing **y** to a negaitive **y**. **Plot 2** below demonstrates the best fit line.

```python
# Plot points.
plt.plot(x, -y, '.')

# Plot line.
plt.plot(xbestfit, ybestfit);
plt.title('Plot 2: Overall Plot Best Fit line', color = "blue");
```

### <ins>Colours</ins>

Here, I changed the **ys** values to negative, so that as **x** goes up **y** goes down. **Plot 3** below shows this with the breakdown of the data into categories using different colour dots. 

```python
# Plot the same points, but now with a different colour for each range.
for i in range(len(xs)):
    plt.plot(xs[i], -ys[i], '.', label=f'{i}')
    plt.legend()
    plt.title('Plot 3: Overall Plot with breakdown of data categories', color = "blue");
```

Again, below, changing the **ys** to negative, made the overall plot inversely proportional (as x goes up y goes down) and the individual groups (or categories) are positively proportional (as x goes up, so too does y). **Plot 4** below shows this. 

```python
# Same plot but with best fit lines for each category.
for i in range(len(xs)):
    # Plot this category.
    plt.plot(xs[i], -ys[i], '.', label=f'{i}')
    # Fit line to category.
    m, c = np.polyfit(xs[i], -ys[i], 1)
    # Create steadily increasing x range for plot.
    xbestfit = np.linspace(xs[i].min(), xs[i].max(), 50)
    # Corresponding y values based on best fit line.
    ybestfit = m * xbestfit + c
    # Plot best fit line for category.
    plt.plot(xbestfit, ybestfit, color='lightgrey')
    plt.legend()
```

### <font color=darkred>Conclusion</font>
***

Having changed the the variable **y** to negative, this made the overall relationship between **x** and **y** an inverse relationship. However, due to Simpsons Paradox it can be seen that the values for **x** and **y** for the individual categories now have a positive relationship.
